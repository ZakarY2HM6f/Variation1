Researchers measured the span, in millimetres, from one eye to the other of 193 male flies. The data are in the file `Datasets/stalkies.csv`.

1. Import the pandas and seaborn modules.
2. Read in the csv file `Datasets/stalkies.csv` into a pandas DataFrame called `stalkies`. 
3. Plot a histogram of `stalkies['span']` and add some useful annotation.

```python
import pandas as pd
import seaborn as sns

stalkies = pd.read_csv('Datasets/stalkies.csv')

g = sns.displot(stalkies['span'])

g.ax.set_xlabel('Eye span (mm)')
g.ax.set_ylabel('Number of flies')
g.ax.set_title('Eye span of 193 male stalk-eyed flies');
```

### Exercise 3.3.3

1. Using pandas, calculate the sample size, mean and standard deviation of eye span to the appropriate number of decimal places.
2. Were your estimates of the mean and standard deviation close to the actual values?

<div class="alert alert-info">
Your answers should be 

    sample size is 193 flies
    mean eye span is 8.56 mm
    st. dev. is 0.40 mm
</div>

```python
n = stalkies['span'].count()    # The sample size
xbar = stalkies['span'].mean()  # The sample mean
s = stalkies['span'].std()      # The sample standard deviation

print(f'sample size is {n} flies')
print(f'mean eye span is {xbar:.2f} mm')
print(f'st. dev. is {s:.2f} mm')
```

### Exercise 3.3.4

If eye span is normally distributed then 68% of the flies' eye spans lie within one standard deviation of the mean.

Using Python code, calculate how many flies you theoretically expect to have eye spans within one standard deviation of the mean if eye span is normally distributed?

<div class="alert alert-info">
Your answer should be 131 flies.
</div>

```python
0.68 * n
```

### Exercise 3.3.5

Using Python code calculate the mean plus one standard deviation and the mean minus one standard deviation.

<div class="alert alert-info">
Your answers should be 

    mean minus one st. dev. = 8.16 mm
    mean plus one st. dev. = 8.95 mm
</div>

```python
print( f'mean minus one st. dev. = {xbar - s:.2f} mm' )
print( f'mean plus one st. dev. = {xbar + s:.2f} mm' )
```

### Exercise 3.3.6

Using Python code calculate how many flies actually have eye spans within one standard deviation of the mean. 

Remember not to hard-code your previous numerical answers.

<div class="alert alert-info">
Your answer should be 133 flies.
</div>

```python
# Set tally of eye spans within one standard deviation of the mean to zero.
count = sum((stalkies['span'] > xbar - s) & (stalkies['span'] < xbar + s))

print( f'{count} stalkies have eye spans between {xbar - s:.2f} mm and {xbar + s:.2f} mm' )
```

2. What proportion of women have heart rates between 55 bpm and 75 bpm?
