Researchers recorded the foraging lifespan in hours of 33 worker bees. 

1. Import the pandas and seaborn modules.
2. Read in the csv file `Datasets/bee lifespans.csv` into a pandas DataFrame called `bees`. 
3. Plot a histogram of `bees['hours']` and add some useful annotation.

```python
import pandas as pd
import seaborn as sns

bees = pd.read_csv('Datasets/bee lifespans.csv')

g = sns.displot(bees['hours'])

g.ax.set_xlabel('Foraging lifespan (hour)')
g.ax.set_ylabel('Number of bees')
g.ax.set_title('Foraging lifespan of 33 worker bees');
```

### Exercise 3.2.2

Using pandas, calculate the standard deviation of foraging lifespan of the bees to the appropriate number of decimal places.

<div class="alert alert-info">
Your answer should be 20.56 hours.
</div>

```python
s = bees.std()

print(f"standard deviation of foraging lifespan is {s['hours']:.2f} hours")
```

Elephant seals defend harems that may contain hundreds of reproductively active females. Researchers recorded the numbers of females in harems in a population on South Georgia Island. 

1. Import the pandas module and seaborn modules.
2. Read in the csv file `Datasets/elephant seal.csv` into a pandas DataFrame called `seal`. 
3. Plot a histogram of `seal['Harem size']` and add some useful annotation.

```python
import pandas as pd
import seaborn as sns

seal = pd.read_csv('Datasets/elephant seal.csv')

g = sns.displot(seal['Harem size'])

g.ax.set_ylabel('Number of harems')
g.ax.set_title('Harem size of elephant seals on South Georgia Island');
```

### Exercise 3.2.4

Using pandas, calculate the standard deviation of harem size to the appropriate number of decimal places.

<div class="alert alert-info">
Your answer should be 41.4 females.
</div>

```python
s = seal.std()

print(f"standard deviation harem size is {s['Harem size']:.1f}")
```
