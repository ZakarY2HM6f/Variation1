One of the commonest uses of inferential statistics is to estimate the population mean using a sample drawn from the population.

We'll begin with how to estimate the population mean, which is very straight forward. Then we'll discuss how precise that estimate is, which starts to get a little more complicated.

We'll use beak depth data of medium ground finches on the Galápagos island of Daphne Major in 1976 to illustrate these ideas.

## Beak depths of the medium ground finch of the Galápagos islands

In 1976 and again in 1978 researchers measured the beak depths of medium ground finches on the Galápagos island of Daphne Major. The data are in the file `Datasets/finches beak depth.csv`.
<br>

<div>

</div>

We'll just examine the 1976 data for now. Next week we'll compare the 1976 and 1978 data.

Let's first read in the data and plot a histogram to see what it looks like. 

<div class="alert alert-info">

Run the code cell below to read in and plot a histogram of data. 
</div>

```python
%matplotlib inline
import pandas as pd
import seaborn as sns

# Read in the medium ground finches beak depth data.
beak_depths = pd.read_csv('Datasets/finches beak depth.csv')

# Get the sample size
n = beak_depths['1976'].count()

# Plot a histogram of the data.
g = sns.displot(beak_depths['1976'])

# Add some annotation.
g.ax.set_xlabel('Beak depth (mm)')
g.ax.set_ylabel('Number of finches')
g.ax.set_title(f"Beak depths of {n} medium ground\nfinches on Daphne Major in 1976");
```

Beak depths vary from about 7 mm to 12 mm with an average of roughly 9.5 mm. The standard deviation is approximately (12 $-$ 7) / 4 = 1.3 mm.

## The estimate of the population mean

As all we have is one sample then the estimate of the population mean is the sample mean. 

<div class="alert alert-info">

Run the code below to calculate the sample mean to 2dp.
</div>

```python
# sample mean
xbar = beak_depths['1976'].mean() 

print(f"sample mean = {xbar:.2f} mm")
```

The mean beak depth of our sample is $\bar{x}$ = 9.49 mm to 2dp. 

**The estimate of the population mean is simply the sample mean. That is, we estimate the population mean to be $\hat{\mu}$ = 9.49 mm.**

Remember that $\mu$ is the population mean, which we do not know, would like to know but never will. $\bar{x}$ is the mean of the sample we took from the population, which we do know. And $\hat{\mu}$ is our estimate of the population mean which equals $\bar{x}$.

## But how precise is the estimate of the population mean?

So the estimate of the population mean is 9.49 mm. But can we tell how precise that estimate is? Are we way out or are we quite close to the actual population mean? Let's do a little thought-experiment to demonstrate what we mean by this question.

Hypothetically, say we sampled just two finches (*n*=2) whose beak depths were 8.6 mm and 10.8 mm. This sample's mean is 9.68 mm ([8.6+10.8]/2), so our estimate of the population mean is 9.68 mm. This is shown in the first row (sample 1) in the Table below. Now let's sample another two finches. Their beak depths are 7.1 mm and 10.0 mm. This sample's mean is 8.52 mm, so our estimate of the population mean is 8.52 mm. We could, in theory, keep taking different samples of two finches forever, each time estimating the population mean.

The following table lists the samples and the estimates of the first four of this infinite set of hypotheitical samples.

Sample | sample size *n* | 1<sup>st</sup> finch | 2<sup>nd</sup> finch | estimate, $\hat{\mu}$, mm
:--- | :---: | ---: | ---: | ---:
sample 1 | 2 | 8.6 | 10.8 | 9.68
sample 2 | 2 | 7.1 | 10.0 | 8.52
sample 3 | 2 | 9.8 | 11.8 | 10.83
sample 4 | 2 | 10.4 | 8.1 | 9.24
and so on to infinity | ... | ... | ... | ... 

Notice how variable the estimates of the population means are, varying from 8.52 mm to 10.83 mm.

Now let's do the same thought-experiment but this time sampling 75 finches each time. The following table lists samples and the estimates of the first four of this infinite set of hypothetical samples. (We don't list beak depths of all 75 finches.)

Sample | sample size *n* | estimate, $\hat{\mu}$, mm
:--- | :---: | ---:
sample 1 | 75 | 9.49
sample 2 | 75 | 9.62
sample 3 | 75 | 9.67
sample 4 | 75 | 9.41
and so on to infinity | ... | ... 

Notice how less variable the estimates of the population means are, varying from 9.41 mm to 9.67 mm.

The estimates are much less variable when we sample 75 finches compared to when we sample two finches. This should come as no surprise. It is intuitive that to get ever more precise estimates of a population mean we need to measure ever more larger samples taken from the population.

In practice we don't want to be doing these thought experiments. Moreover, in practice we only ever have one sample. We want some measure, some number, that tells us how precise our estimate of the population mean is given our single sample. To answer this question we need to take a detour into something called a sampling distribution.

## A histogram of estimates of the population mean: The sampling distribution

Consider our infinite lists of estimates of hypothetical samples of size 2 and size 75 that we tabulated above. We can plot histograms of both of these tables of estimates like so:



The histogram on the left shows samples of size 2 and the histogram on the right shows samples of size 75. This clearly demonstrates that larger sample sizes result in more precise estimates of the population mean.

A histogram of sample means has a special name: the **sampling distribution of the mean**. Sampling distributions are of fundamental importance in inferential statistics. Understanding sampling distributions will help you immensely in understanding statistics.

Now, our actual estimate of the 1976 population mean is 9.49 mm which we calculated from the sample of 75 finches. This estimate is one of the infinite number of estimates that make up the sampling distribution on the right.

The width of this sampling distribution tells us how variable our estimates of the population mean would be **if** we could sample the population of finches over and over again. **So the width of the sampling distribution tells us how precise our actual estimate of the population mean is.**

We can use the standard deviation of the sampling distribution as a measure of its width. Remember standard deviation is just a way of measuring how spread out (how variable) something is. The standard deviation of the sampling distribution on the right turns out to be 0.11 mm. 

It's also important you understand the difference between a **sample** distribution and a **sampling** distribution. A sample distribution is a histogram of a sample of data collected from an experimental or observational study. For example, the histogram of 75 ground finch beak depths collected from Daphne Major in 1976 which you plotted above is a sample distribution. A sampling distribution, on the other hand, is a hypothetical histogram of the means of an infinite number of samples of size *n* drawn from the population. 

Before explaining how we use the standard deviation to report the precision of our estimate we need to introduce some more jargon that will be useful later on.

## Standard error of the mean

The standard deviation of a sampling distribution has a special name: the **standard error** (**SE** for short). The standard deviation of the sampling distribution of the mean is called the **standard error of the mean** (**SEM** for short). (As the mean is usually the most important characteristic of a set of data, people often use "standard error" to mean "standard error of the mean".) 

The formula for calculating the the standard error of the mean is simply

$$ \mathrm{SEM} \approx \frac{s}{\sqrt{n}}$$

The squiggly equals sign, $\approx$, means "approximately equal to". The standard error is approximately equal to the standard deviation of the sample divided by the square root of the sample size. The derivation of this formula is not important. But what is important is that the sample size occurs in the denominator (the bottom) of this formula. What this tells us is that as we increase the sample size, the standard error gets smaller - the sampling distribution gets narrower - as we saw above. 

We can now calculate the standard error of the mean for the beak depth data.

<div class="alert alert-info">

1. Read the code below to make sure you know how the standard error of the mean is calculated
2. Run the code
</div>

```python
import math  # Import math for the square root function

n = beak_depths['1976'].count()   # sample size
s = beak_depths['1976'].std()     # sample standard deviation

SEM = s / math.sqrt(n)   # the standard error of the mean for the 1976 finches

print(f'n = {n} finches')
print(f's = {s:.2f} mm')
print()
print(f'SEM = {SEM:.2f} mm')
```

What does this mean? It means that with 75 finches we are able to able to estimate the population mean beak depth to precision of a tenth of a millimetre. If you think about it that's quite remarkable: beak depths varied from 7 mm to 12 mm - a range of 5 mm. But by measuring the beak depths of 75 finches, the precision of the estimate of the population mean is a tiny 0.1 mm, fifty times smaller!

Another important insight about sampling is that while larger samples are always better (in terms of their ability to accurately represent the entire population), there are diminishing returns as the sample gets larger. In fact, the rate at which the benefit of larger samples decreases follows a simple mathematical rule, growing as the square root of the sample size, such that in order to double the precision of our estimate we need to quadruple the size of our sample.
